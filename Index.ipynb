```python
import json
import ast
import pandas as pd
import urllib
import requests
```

```python
r = urllib.request.urlopen('https://servicodados.ibge.gov.br/api/v2/censos/nomes/joao%7Cenzo')
data = json.loads(r.read())
```

```python
enzo = data[0]
joao = data[1]
```

```python
x = (pd.DataFrame.from_dict(enzo['res'],))
y = (pd.DataFrame.from_dict(joao['res'],))
```

```python
import matplotlib.pyplot as plt
```

```python
period = x['periodo']
frenzo = x['frequencia']
frjoao = y['frequencia']
plt.plot(period, frenzo, color='blue')
plt.xlabel('Períodos')
plt.ylabel('Frequência')
plt.title('Frequência de nomes no Brasil por década')
plt.show()
plt.plot(period, frjoao, color='orange')
plt.xlabel('Períodos')
plt.ylabel('Frequência')
plt.title('Frequência de nomes no Brasil por década')
plt.show()
```
